### Function to encode categorical values 

```python
from sklearn.preprocessing import LabelEncoder
def encode_categorical(df, cols):
    for col in cols :
        le = LabelEncoder()
        
        new_col = col+"_num"
        df[new_col] = le.fit_transform(df[col])

    return df

cols = ['Region', 'Soil_Type', 'Weather_Condition']
df = encode_categorical(df,cols)
```

### Checking content of dataframe after each transfromation

```python
df
```

###  Adding new columns that can be useful

```python
df['Yield_kg_per_sqm'] = df['Yield_tons_per_hectare'] * 1000 / 10000
```

### Reformatting all float data to two decimal places

```python
df = df.map(lambda x: f"{x:.2f}" if isinstance(x, float) else x)
```

### Method for Scaling columns you want

```python
from sklearn.preprocessing import MinMaxScaler

def scale_columns(df, columns):
    scaler = MinMaxScaler()

    scaled_data = scaler.fit_transform(df[columns])

    scaled_column_names = [f"{col}_scaled" for col in columns]

    df[scaled_column_names] = scaled_data

    return df

df = scale_columns(df, ['Yield_tons_per_hectare', 'Rainfall_mm', 'Yield_kg_per_sqm'])
```

### Loading dataframe to Postgresql Database

## Q1
### Average Yield by Region and Soil Type

```python
query1 = """
SELECT 
    "Region",
    "Soil_Type",
    AVG("Yield_tons_per_hectare"::numeric) AS avg_yield_tons_per_hectare
FROM 
    crop_yield
GROUP BY 
    "Region", "Soil_Type"
ORDER BY 
    avg_yield_tons_per_hectare DESC;
"""

df_avg_yield = pd.read_sql(query1, engine)
print("Average Yield by Region and Soil Type:")
print(df_avg_yield)
```

## Q2
### High-Yield Crops in Rainy Conditions

```python
query2 = """
SELECT 
    "Crop",
    "Region",
    "Yield_tons_per_hectare"::numeric,
    "Weather_Condition"
FROM 
    crop_yield
WHERE 
    "Weather_Condition" = 'Rainy' 
    AND "Yield_tons_per_hectare"::numeric > 7
ORDER BY 
    "Yield_tons_per_hectare" DESC;
"""

df_high_yield_rainy = pd.read_sql(query2, engine)
print("High-Yield Crops in Rainy Conditions:")
print(df_high_yield_rainy)
```

## Q3
### Scaled vs Actual Rainfall for Crops in South Region

```python
query3 = """
SELECT 
    "Region",
    "Crop",
    "Rainfall_mm"::numeric,
    "Rainfall_mm_scaled"::numeric
FROM 
    crop_yield
WHERE 
    "Region" = 'South'
LIMIT 10;
"""

df_scaled_rainfall = pd.read_sql(query3, engine)
print("Scaled vs Actual Rainfall for Crops in South Region:")
print(df_scaled_rainfall)
```

## Resulting Table description

| Original Column                    | Description                                                                                           |
|---------------------------|-------------------------------------------------------------------------------------------------------|
| **Region**                | The geographical area where the crop is grown (e.g., North, South, West).                             |
| **Soil_Type**             | Type of soil in the region where the crop is cultivated (e.g., Sandy, Clay, Loam, Silt). This is important for understanding the soil's impact on crop yield. |
| **Crop**                  | Type of crop being cultivated (e.g., Cotton, Rice, Barley, Soybean, Wheat). Each crop has different growth requirements and yields. |
| **Rainfall_mm**           | The amount of rainfall in millimeters during the crop-growing season. Rainfall influences crop growth and water availability. |
| **Temperature_Celsius**   | The average temperature in degrees Celsius during the crop-growing period. Temperature affects the rate of crop growth and potential yield. |
| **Fertilizer_Used**       | Indicates whether fertilizer was applied (`True` for yes, `False` for no). Fertilizer can improve soil nutrients and impact yield. |
| **Irrigation_Used**       | Indicates whether irrigation was used (`True` for yes, `False` for no). Irrigation provides water to crops and can impact yield, especially in low-rainfall areas. |
| **Weather_Condition**     | General weather conditions during the crop-growing season (e.g., Cloudy, Rainy, Sunny). This can impact sunlight availability and overall crop health. |
| **Days_to_Harvest**       | The number of days required for the crop to reach harvest maturity. Different crops and conditions will have varying growth times. |




| New Column                    | Description                                                                                           |
|---------------------------|-------------------------------------------------------------------------------------------------------|
| **Yield_tons_per_hectare** | Crop yield measured in tons per hectare. This metric represents the amount of crop produced per unit area, useful for comparing crop productivity. |
| **Region_num**            | Numeric encoding of the `Region` column, useful for machine learning models that require numeric inputs. |
| **Soil_Type_num**         | Numeric encoding of the `Soil_Type` column, representing each soil type as a unique number for analysis purposes. |
| **Weather_Condition_num** | Numeric encoding of the `Weather_Condition` column, converting weather types into numerical categories. |
| **Yield_kg_per_sqm**      | Yield measured in kilograms per square meter, which is another way to express crop productivity at a smaller unit scale. |
| **Yield_tons_per_hectare_scaled** | Scaled version of `Yield_tons_per_hectare`, normalized to a range (likely 0-1) to make data comparable across different columns. |
| **Rainfall_mm_scaled**    | Scaled version of `Rainfall_mm`, normalized for easier analysis and to remove the effects of units and magnitude differences. |
| **Yield_kg_per_sqm_scaled** | Scaled version of `Yield_kg_per_sqm`, making it comparable with other scaled features. |

This dataset provides comprehensive information on crop productivity influenced by environmental and agronomic factors, which could be useful in predictive modeling or agricultural studies.
